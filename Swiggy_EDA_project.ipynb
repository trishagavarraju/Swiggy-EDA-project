```python
#which city we are getting most orders from:
city_counts=df["City"].value_counts()
print(city_counts)
plt.figure()
plt.bar(city_counts.index,city_counts.values)
plt.title("total orders per city")
plt.xticks(rotation=75)
plt.show()
```

```python
plt.figure(figsize=(11,6))
plt.pie(city_counts.values,labels=city_counts.index,autopct="%.2f%%")
plt.title("total orders per city")
plt.show()
```
